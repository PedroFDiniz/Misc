```python
import numpy as np
from sklearn.cluster import KMeans
from collections import Counter
import imutils
import cv2
import pprint
from matplotlib import pyplot as plt
from math import sqrt,pow

MAX_EUCLIDEAN_DISTANCE = 441.672955930063709849498817084

def getColorInformation(labels, clusters):
  
  # Contador de ocorrências de cada cor estimada (predita) pelo kmeans
  occurence_counter = Counter(labels)
  print(occurence_counter)
  
  # Array de dicionários de cores
  colorInformation = []
 
  # Obtém a soma total de todas as ocorrências preditas
  totalOccurance = sum(occurence_counter.values()) 
  
  # Percorre todas as cores dominantes (mais frequentes) preditas
  for x in occurence_counter.most_common(len(clusters)):
    
    index = (int(x[0]))
    
    # Converte o número da cor em uma lista de 3 componentes (RGB)
    color = clusters[index].tolist()
    
    # Obtém a porcentagem da cada cor
    color_percentage= (x[1]/totalOccurance)
    
    # Constrói o dicionário que mapeia índices de cores em cores e porcentagens
    colorInfo = {"cluster_index":index , "color": color , "color_percentage" : color_percentage }
    
    # Acrescenta o dicionário à lista de dicionários de cores
    colorInformation.append(colorInfo)
    
  return colorInformation

def extractDominantColor(image,number_of_colors=5):
    
  # Cria uma cópia da imagem
  img = image.copy()
  
  # Reorganiza a imagem para que que seja um longo array unidimensional de RGBs
  img = img.reshape((img.shape[0]*img.shape[1]) , 3)
  
  #Inicializa o objeto KMeans, informando o número de clusters desejado
  clustering = KMeans(n_clusters=number_of_colors, random_state=0)
  
  # Realiza o agrupamento
  clustering.fit(img)
  
  # Obtém a informação das cores dominantes a partir dos rótulos e centróides dos clusters encontrados
  colorInformation = getColorInformation(clustering.labels_, clustering.cluster_centers_)
  
  return colorInformation

def plotColorBar(colorInformation):
  #Create a 500x100 black image
  color_bar = np.zeros((100,500,3), dtype="uint8")
  
  top_x = 0
  for x in colorInformation:    
    bottom_x = top_x + (x["color_percentage"] * color_bar.shape[1])

    color = tuple(map(int,(x['color'])))
  
    cv2.rectangle(color_bar , (int(top_x),0) , (int(bottom_x),color_bar.shape[0]) ,color , -1)
    top_x = bottom_x
  return color_bar

def prety_print_data(color_info):
  for x in color_info:
    print(pprint.pformat(x))
    print()

images = ["https://img.freepik.com/fotos-gratis/bela-paisagem-de-um-rio-cercado-por-muito-verde-em-uma-floresta_181624-40482.jpg?w=1380&t=st=1682284278~exp=1682284878~hmac=44aac5e14e48bd610c4ba892eb15e7d37c1b4b610d916736c4306ad849cc4713",
        "https://img.freepik.com/fotos-gratis/linda-arvore-no-meio-de-um-campo-coberto-de-grama-com-a-linha-das-arvores-ao-fundo_181624-29267.jpg?w=1480&t=st=1682284321~exp=1682284921~hmac=abe082969234048d3271caba11bd1452fb9e10b629614d0a8e506ef9ddac7d52",
        "https://img.freepik.com/fotos-gratis/caminho-de-terra-em-um-parque-florestal-passando-por-uma-arca-de-pedra-na-serra-do-bucaco-portugal_181624-34388.jpg?w=1380&t=st=1682284352~exp=1682284952~hmac=3c5abe0fdafbf9ed7ca45de2f4109b35ac2fe78d1c6b9370d4d9282e034312af",
        "https://img.freepik.com/fotos-gratis/campo-com-grama-e-nuvens_1112-621.jpg?w=1380&t=st=1682296923~exp=1682297523~hmac=60b1ef42c64991724d437f33fc18b1c952dde99f14f669abed0bc60febe762e4",
        "https://img.freepik.com/fotos-gratis/nascer-do-sol-sobre-a-selva-de-bali_1385-1644.jpg?w=1380&t=st=1682284384~exp=1682284984~hmac=1d921e8ba79b00d213193b8ee91c6796d90df32bb51a2b51710c15d808be5a59",
        "https://img.freepik.com/fotos-gratis/agua-da-chuva-em-um-macro-verde-da-folha_1150-12802.jpg?w=1380&t=st=1682284411~exp=1682285011~hmac=361eabefac575cba871cafa2d21e459c1e583b285fa06d203e77af652b7968e1",
        "https://img.freepik.com/fotos-premium/folha-da-folha-tropical-na-obscuridade-verde-com-gota-da-agua-de-chuva-na-textura-fundo-abstrato-da-natureza-do-teste-padrao_1484-2338.jpg?w=1380",
        "https://img.freepik.com/fotos-gratis/mulher-na-ponta-da-piedade-um-local-pitoresco-em-portugal_181624-35807.jpg?w=1380&t=st=1682290704~exp=1682291304~hmac=6bed05f3515696ec72a71ee37c32e5f06277efd48f32db816a44506978521b74",
        "https://img.freepik.com/fotos-gratis/bela-paisagem-natural-com-arco-iris_23-2150169060.jpg?w=1380&t=st=1682296784~exp=1682297384~hmac=a4779a32f0d006662c36696c443ad33bbba990ce2ad253bbfbde972cbf62ed7b",
        "https://img.freepik.com/fotos-gratis/floresta-de-inverno-de-manha-cedo-ao-nascer-do-sol_169016-29714.jpg?w=740&t=st=1682296828~exp=1682297428~hmac=28da80da45255582d8588865df7a92451235c4677eb5811e626a464d7fd3fe18",
        "https://img.freepik.com/fotos-gratis/barco-em-um-lago-ao-por-do-sol_181624-40928.jpg?w=1380&t=st=1682296850~exp=1682297450~hmac=9a808f82c9c808cf0044ac0a747028bee421ae8a7dc804cdb172d065c96d065d",
        "https://img.freepik.com/fotos-gratis/foto-de-tirar-o-folego-da-ponte-vasco-da-gama-ao-nascer-do-sol-em-lisboa-portugal_181624-46616.jpg?w=1380&t=st=1682296881~exp=1682297481~hmac=f69901f94581fc23cc95a2b944a91acab28ed1a531a502d0595b28dd6b9ef464"]

imgs_info=[]
# Get Image from URL. If you want to upload an image file and use that comment the below code and replace with  image=cv2.imread("FILE_NAME")
for imgstr in images:
  image =  imutils.url_to_image(imgstr)

  # Resize image to a width of 250
  image = imutils.resize(image,width=250)
  image = cv2.cvtColor(image,cv2.COLOR_BGR2RGB)

  #Show image
  plt.imshow(image)
  plt.show()


  # Find the dominant color. Default is 1 , pass the parameter 'number_of_colors=N' where N is the specified number of colors 
  dominantColors = extractDominantColor(image, number_of_colors=5)
  imgs_info.append(dominantColors)

  #Show in the dominant color as bar
  print("Color Bar")
  colour_bar = plotColorBar(dominantColors)
  plt.axis("off")
  plt.imshow(colour_bar)
  plt.show()

  #Show in the dominant color information
  print("Color Information")
  prety_print_data(dominantColors)



def compare_colors(img1,img2):
  result = []
  for i in range(5):
    #Usando os valores RGB como coordenadas Euclidianas
    dist_red = img1[i]['color'][0]-img2[i]['color'][0]
    dist_grn = img1[i]['color'][1]-img2[i]['color'][1]
    dist_blu = img1[i]['color'][2]-img2[i]['color'][2]
    dist = sqrt(pow(dist_red,2) + pow(dist_grn,2) + pow(dist_blu,2))
    color_simil = 100 - ((dist/MAX_EUCLIDEAN_DISTANCE)*100)
    degree_diff = (img1[i]['color_percentage']-img2[i]['color_percentage'])
    degree_pctg = 100-sqrt(pow(degree_diff*100,2))
    result.append(tuple((i,i,dist,color_simil,degree_pctg)))
  return result

#Compara a imagem cujo índice é passado como parâmetro a todas as outras imagens da lista
def compare_to_this(indice):
  similarities = []
  comparison_values = []
  for i in range(len(imgs_info)):
    if i != indice:
      similarities.append(compare_colors(imgs_info[indice],imgs_info[i]))
  for j in range(len(similarities)):
    media_clr = 0
    media_pct = 0
    for t in range(len(similarities[j])):
#      pctg = "{:.2f}".format(similarities[j][t][3])
#      dist = "{:.2f}".format(similarities[j][t][2])
      media_clr += similarities[j][t][3]
      media_pct += similarities[j][t][4]
      clr = str(similarities[j][t][0] + 1)
#      print("Cor " + clr + " com distancia " + dist + ", equivalente a " + pctg + "% de similaridade de cor")
    media_clr = media_clr/len(similarities[j])
    media_pct = media_pct/len(similarities[j])
    comparison_values.append(tuple((media_clr,media_pct)))
  return comparison_values
#    pctg = "{:.2f}".format(media_clr)
#    clr_pctg = "{:.2f}".format(media_pct)
#    print("Média de " + pctg + "% de similaridade de cores nas imagens")
#    print("Similaridade de porcentagem de " + clr_pctg + "%")
#    print()

compare_to_this(0)
```
